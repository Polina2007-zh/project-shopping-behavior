# Обработка данных

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype in ["int64", "float64"]:
            df[col] = df[col].fillna(df[col].median())
        else:
            df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
```

```python
z_scores = np.abs(
    (df[numeric_cols] - df[numeric_cols].mean()) / df[numeric_cols].std()
)
```

```python
(z_scores > 3).sum()
```

# Группировки и агрегации

```python
df.groupby("Category").agg(
    mean_value=("Purchase Amount (USD)", "mean"),
    median_value=("Purchase Amount (USD)", "median"),
    count_value=("Purchase Amount (USD)", "count"),
    sum_value=("Purchase Amount (USD)", "sum")
)
```

```python
pd.pivot_table(
    df,
    values="Purchase Amount (USD)",
    index="Category",
    aggfunc=["mean", "count"]
)
```
